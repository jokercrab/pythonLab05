1. Загрузити файл з даними про споживання енергії “En_In.xls”, який являє
собою перелік показників енергозабезпечення та виробництва відновлюваної
електроенергії, і ввести їх в DataFrame.

```python
data=pd.read_excel("En_In.xls",names=['1','2','Country','EnergySup','EnergySupprCap','%Renewable'])
data=data[17:244][['Country','EnergySup','EnergySupprCap','%Renewable']]

data
```

```python
mask = data[['EnergySup','EnergySupprCap','%Renewable']].applymap(lambda x: isinstance(x, (int, float)))
data[['EnergySup','EnergySupprCap','%Renewable']]=data[['EnergySup','EnergySupprCap','%Renewable']].where(mask)
data
```

3. Існує декілька країн, що мають в назві цифри та/або дужки. Обов’язково
видаліть їх, напр. 'Bolivia (Plurinational State of)' повинна бути
‘Bolivia’, ‘Switzerland17’ повинна бути Switzerland’.

```python
data.Country=data['Country'].str.replace(r" \(.*\)","",regex=True)
data.Country=data['Country'].str.replace(r"\d+","",regex=True)
```
